```python
import pandas as pd 
###
```

```python
# Load the first dataset (assuming it's in a CSV file named 'weather_data.csv')
df1 = pd.read_csv('data/weather-data.csv', parse_dates=['time'])

# Load the second dataset (assuming it's in a CSV file named 'air_quality_data.csv')
df2 = pd.read_csv('data/train_with_holidays.csv', parse_dates=['id'])
```

```python
# For df1, the time is already in the correct format
df1.set_index('time', inplace=True)

# For df2, convert the 'id' column to datetime and set it as the index
df2['id'] = pd.to_datetime(df2['id'])
df2.set_index('id', inplace=True)
```

```python
# Merge the datasets on their indices (which are now both in datetime format)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
```

```python
# Save the merged dataset to a new CSV file
merged_df.to_csv('data/merged_data.csv')
```
